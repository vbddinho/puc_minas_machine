```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Definição de configurações de dataframes
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.set_option('display.precision', 4)
pd.set_option('display.expand_frame_repr', False)
```

```python
#Configurando os plots para serem emandanteibidos diretamente no notebook
%matplotlib inline
```

```python
df_jogos = pd.read_csv('dados/campeonato-brasileiro-full.csv',sep=',',encoding = "UTF-8")
df_estatisticas = pd.read_csv('dados/campeonato-brasileiro-estatisticas-full.csv', sep=',',encoding = "UTF-8")
```

```python
df_jogos.info()
```

```python
df_jogos.rename(columns={'ID':'partida_id'},inplace=True)
df_jogos.rename(columns={'rodata':'rodada'},inplace=True)
df_jogos.drop(columns=['hora','arena','visitante_Estado','mandante_Estado','tecnico_mandante','tecnico_visitante'],inplace=True)
```

```python
df_estatisticas.tail()
```

```python
df_jogos.tail()
```

```python
#Fonte https://pt.stackoverflow.com/questions/484866/como-preencher-uma-coluna-de-um-df-pandas-usando-como-compara%C3%A7%C3%A3o-uma-coluna-es
df_jogos = pd.merge(df_jogos,df_estatisticas,how='left',right_on=['partida_id','clube'],left_on=['partida_id','mandante'])
df_jogos = pd.merge(df_jogos,df_estatisticas,how='left',right_on=['partida_id','clube'],left_on=['partida_id','visitante'])
```

```python
df_jogos.columns
```

```python
df_jogos.columns = ['partida_id', 'rodada', 'data', 'mandante', 'visitante', 'formacao_mandante', 'formacao_visitante', 'vencedor', 'mandante_placar', 'visitante_placar', 'rodata_mandante', 'clube_mandante', 'chutes_mandante', 'chutes_no_alvo_mandante', 'posse_de_bola_mandante', 'passes_mandante', 'precisao_passes_mandante', 'faltas_mandante', 'cartao_amarelo_mandante', 'cartao_vermelho_mandante', 'impedimentos_mandante', 'escanteios_mandante', 'rodata_visitante', 'clube_visitante', 'chutes_visitante', 'chutes_no_alvo_visitante', 'posse_de_bola_visitante', 'passes_visitante', 'precisao_passes_visitante', 'faltas_visitante', 'cartao_amarelo_visitante', 'cartao_vermelho_visitante', 'impedimentos_visitante', 'escanteios_visitante']
```

```python
df_jogos.drop(columns=['rodata_mandante','mandante','visitante','rodata_visitante'],inplace=True)
```

```python
df_jogos.tail()
```

```python
#Converte a data do tipo objeto para o tipo datetime e cria a coluna ano
df_jogos["data"] = pd.to_datetime(df_jogos["data"],format='%d/%m/%Y',dayfirst=True)
df_jogos["ano"] = df_jogos["data"].dt.year
```

```python
#temporada de 2020 e 2021 foi afetada pelo COVID-19, realizando ajuste
df_jogos.loc[(df_jogos["data"]>="2020/08/08") & (df_jogos["data"]<="2021/02/25"),"ano" ] = 2020
df_jogos.loc[(df_jogos["data"]>="2021/05/29") & (df_jogos["data"]<="2021/12/09"),"ano" ] = 2021
df_jogos.rename(columns={'ano':'temporada'},inplace=True)
```

```python
df_jogos.shape
```

```python
df = df_jogos[df_jogos['temporada']>=2018]

df_jogos[df_jogos['temporada']>=2018].isna().sum()
```

```python
df.tail()
```

```python
df.loc[(df['vencedor']==df['clube_mandante']),'mandante_win'] = 1
df.loc[(df['vencedor']==df['clube_visitante']),'visitante_win'] = 1
df.loc[(df['vencedor']=='-'),'empate'] = 1

df.loc[(df['vencedor']==df['clube_mandante']),'resultado'] = 'Vencedor-Mandante'
df.loc[(df['vencedor']==df['clube_visitante']),'resultado'] = 'Vencedor-Visitante'
df.loc[(df['vencedor']=='-'),'resultado'] = 'Empate'

df_jogos.drop(columns=['vencedor'],inplace=True)
```

```python
df['mandante_win'].fillna(value=0, inplace=True)
df['visitante_win'].fillna(value=0, inplace=True)
df['empate'].fillna(value=0, inplace=True)
```

```python
print(" Quantidade de Partidas analisadas: " , str(df.shape[0]))
```
