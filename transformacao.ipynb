```python
df_jogos = pd.read_csv('/content/campeonato-brasileiro-full.csv',sep=',',encoding = "UTF-8")
df_estatisticas = pd.read_csv('/content/campeonato-brasileiro-estatisticas-full.csv', sep=',',encoding = "UTF-8")
```

```python
#colocando todas as colunas em minusculo
df_jogos.columns = df_jogos.columns.str.lower()
df_estatisticas.columns = df_estatisticas.columns.str.lower()
```

```python
df_jogos.info()
```

```python
df_jogos.rename(columns={'id':'partida_id'},inplace=True)
df_jogos.rename(columns={'rodata':'rodada'},inplace=True)
```

```python
df_estatisticas.info()
```

```python
# left join dos dados de mandante
dfmandante = pd.merge(df_jogos,df_estatisticas,how='left',right_on=['partida_id','clube'],left_on=['partida_id','mandante'])
```

```python
# left join dos dados de mandante
df = pd.merge(dfmandante,df_estatisticas,how='left',right_on=['partida_id','clube'],left_on=['partida_id','visitante'])
```

```python
# renomenando as colunas de mandante e visitante
df.columns = ['partida_id', 'rodada', 'data', 'hora', 'mandante', 'visitante', 'formacao_mandante', 'formacao_visitante', 'tecnico_mandante', 'tecnico_visitante', 'vencedor', 'arena', 'mandante_placar', 'visitante_placar', 'mandante_estado', 'visitante_estado', 'rodata_mandante', 'clube_mandante', 'chutes_mandante', 'chutes_no_alvo_mandante', 'posse_de_bola_mandante', 'passes_mandante', 'precisao_passes_mandante', 'faltas_mandante', 'cartao_amarelo_mandante', 'cartao_vermelho_mandante', 'impedimentos_mandante', 'escanteios_mandante', 'rodata_visitante', 'clube_visitante', 'chutes_visitante', 'chutes_no_alvo_visitante', 'posse_de_bola_visitante', 'passes_visitante', 'precisao_passes_visitante', 'faltas_visitante', 'cartao_amarelo_visitante', 'cartao_vermelho_visitante', 'impedimentos_visitante', 'escanteios_visitante']
```

```python
df.set_index("partida_id",inplace=True,drop=False)
df.tail()
```

```python
#Converte a data do tipo objeto para o tipo datetime e cria a coluna ano
df["data"] = pd.to_datetime(df["data"],format='%d/%m/%Y',dayfirst=True)
```

```python
df["ano"] = df["data"].dt.year
```

```python
df.loc[(df["data"]>="2020/08/08") & (df["data"]<="2021/02/25"),"ano" ] = 2020
df.loc[(df["data"]>="2021/05/29") & (df["data"]<="2021/12/09"),"ano" ] = 2021
df.rename(columns={'ano':'temporada'},inplace=True)
```

```python
#copia do dataframe para o df2, para manter o original df
df2 = df.copy(deep=True)
```

```python
df2 = df[df['temporada']>=2018]

# Verificando que não há dados NAN no dataset
df2.isna().sum()
```

```python
#Definindo colunas de vitorias e derrotas e empates mandante/visitante
df2['resultado'] = df2.apply(lambda x: 1 if x["vencedor"] == x["mandante"] else 2 if x["vencedor"] == x["visitante"] else 3 if x["vencedor"] == '-' else 9999  ,axis=1)
#df['resultado'] = df.apply(lambda x: 2 if x["vencedor"] == x["visitante"] else 0,axis=1)
#df['resultado'] = df.apply(lambda x: 3 if x["vencedor"] == '-' else 5,axis=1)
df2['resultado']
```

```python
#df2018_2022['chutes'] = df2018_2022['chutes'].astype('Int64')

df2['posse_de_bola_mandante'] = df2['posse_de_bola_mandante'].replace({'%': ''}, regex=True)
df2['precisao_passes_mandante'] = df2['precisao_passes_mandante'].replace({'%': ''}, regex=True)
df2['posse_de_bola_visitante'] = df2['posse_de_bola_visitante'].replace({'%': ''}, regex=True)
df2['precisao_passes_visitante'] = df2['precisao_passes_visitante'].replace({'%': ''}, regex=True)
```

```python
# Substitui valores ausentes ou 'None' por NaN
df2['posse_de_bola_mandante'] = df2['posse_de_bola_mandante'].replace({'None': np.nan})
df2['precisao_passes_mandante']= df2['precisao_passes_mandante'].replace({'None': np.nan})
df2['posse_de_bola_visitante'] = df2['posse_de_bola_visitante'].replace({'None': np.nan})
df2['precisao_passes_visitante'] = df2['precisao_passes_visitante'].replace({'None': np.nan})

# convertendo o percentual para numero
df2['posse_de_bola_mandante'] = df2['posse_de_bola_mandante'].astype('Int64') /100
df2['precisao_passes_mandante'] = df2['precisao_passes_mandante'].astype('Int64') /100
df2['posse_de_bola_visitante'] = df2['posse_de_bola_visitante'].astype('Int64') /100
df2['precisao_passes_visitante'] = df2['precisao_passes_visitante'].astype('Int64') /100
```

```python
df2.tail()
```

```python
df2.drop(columns=['data','hora',"tecnico_mandante","tecnico_visitante","vencedor","arena",'mandante_estado','visitante_estado','rodata_mandante'],inplace=True)
```

```python
df2['posse_de_bola_mandante'].fillna(value=0, inplace=True)
df2['precisao_passes_mandante'].fillna(value=0, inplace=True)
df2['posse_de_bola_visitante'].fillna(value=0, inplace=True)      
df2['precisao_passes_visitante'].fillna(value=0, inplace=True) 
```

```python
#Identifiquei que havia valores zerados nos campos abaixo que não fazem sentido, por isso preencho a mediana para os campos zerados.

mpbm =  df2.posse_de_bola_mandante.median()
mpm =  df2.passes_mandante.median()
ppm = df2.precisao_passes_mandante.median()
cm = df2.chutes_mandante.median()
cam = df2.chutes_no_alvo_mandante.median()
cv = df2.chutes_visitante.median()
cav = df2.chutes_no_alvo_visitante.median()
pbv = df2.posse_de_bola_visitante.median()
pv = df2.passes_visitante.median()
ppv = df2.precisao_passes_visitante.median()


df2.posse_de_bola_mandante.replace(0, np.nan, inplace = True)
df2.passes_mandante.replace(0, np.nan, inplace = True)
df2.precisao_passes_mandante.replace(0, np.nan, inplace = True)
df2.chutes_visitante.replace(0, np.nan, inplace = True)
df2.chutes_mandante.replace(0, np.nan, inplace = True)
df2.chutes_no_alvo_visitante.replace(0, np.nan, inplace = True)
df2.chutes_no_alvo_mandante.replace(0, np.nan, inplace = True)
df2.posse_de_bola_visitante.replace(0, np.nan, inplace = True)
df2.passes_visitante.replace(0, np.nan, inplace = True)
df2.precisao_passes_visitante.replace(0, np.nan, inplace = True)


df2.fillna(mpbm,axis=1)
df2.fillna(mpm,axis=1)
df2.fillna(ppm,axis=1)
df2.fillna(cm,axis=1)
df2.fillna(cam,axis=1)
df2.fillna(cv,axis=1)
df2.fillna(cav,axis=1)
df2.fillna(pbv,axis=1)
df2.fillna(pv,axis=1)
df2.fillna(ppv,axis=1)


print("posse_de_bola_mandante: " , mpbm)
print("passes_mandante: " , mpm )
print("precisao_passes_mandante: " , ppm )
print("chutes_mandante: " , cm )
print("chutes_no_alvo_mandante: " , cam )
print("chutes_visitante: " , cv )
print("chutes_no_alvo_visitante: " , cav )
print("posse_de_bola_visitante" , pbv )
print("passes_visitante" , pv )
print("precisao_passes_visitante" , ppv )




#df[(df['passes_mandante']==0)]
```
