```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Definição de configurações de dataframes
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.set_option('display.precision', 4)
pd.set_option('display.expand_frame_repr', False)
```

```python
#Configurando os plots para serem emandanteibidos diretamente no notebook
%matplotlib inline
```

```python
df_jogos = pd.read_csv('dados/campeonato-brasileiro-full.csv',sep=',',encoding = "UTF-8")
df_estatisticas = pd.read_csv('dados/campeonato-brasileiro-estatisticas-full.csv', sep=',',encoding = "UTF-8")

#colocando todas as colunas em minusculo
df_jogos.columns = df_jogos.columns.str.lower()
df_estatisticas.columns = df_estatisticas.columns.str.lower()
```

```python
df_jogos.info()
```

```python
df_jogos.rename(columns={'id':'partida_id'},inplace=True)
df_jogos.rename(columns={'rodata':'rodada'},inplace=True)
```

```python
df_estatisticas.tail()
```

```python
# left join dos dados de mandante
dfmandante = pd.merge(df_jogos,df_estatisticas,how='left',right_on=['partida_id','clube'],left_on=['partida_id','mandante'])
```

```python
# left join dos dados de mandante
df = pd.merge(dfmandante,df_estatisticas,how='left',right_on=['partida_id','clube'],left_on=['partida_id','visitante'])
```

```python
# renomenando as colunas de mandante e visitante
df.columns = ['partida_id', 'rodada', 'data', 'hora', 'mandante', 'visitante', 'formacao_mandante', 'formacao_visitante', 'tecnico_mandante', 'tecnico_visitante', 'vencedor', 'arena', 'mandante_placar', 'visitante_placar', 'mandante_estado', 'visitante_estado', 'rodata_mandante', 'clube_mandante', 'chutes_mandante', 'chutes_no_alvo_mandante', 'posse_de_bola_mandante', 'passes_mandante', 'precisao_passes_mandante', 'faltas_mandante', 'cartao_amarelo_mandante', 'cartao_vermelho_mandante', 'impedimentos_mandante', 'escanteios_mandante', 'rodata_visitante', 'clube_visitante', 'chutes_visitante', 'chutes_no_alvo_visitante', 'posse_de_bola_visitante', 'passes_visitante', 'precisao_passes_visitante', 'faltas_visitante', 'cartao_amarelo_visitante', 'cartao_vermelho_visitante', 'impedimentos_visitante', 'escanteios_visitante']
```

```python
df.set_index("partida_id",inplace=True,drop=False)
df.tail()
```

```python
#Converte a data do tipo objeto para o tipo datetime e cria a coluna ano
df["data"] = pd.to_datetime(df["data"],format='%d/%m/%Y',dayfirst=True)
```

```python
df["ano"] = df["data"].dt.year
```

```python
df.loc[(df["data"]>="2020/08/08") & (df["data"]<="2021/02/25"),"ano" ] = 2020
df.loc[(df["data"]>="2021/05/29") & (df["data"]<="2021/12/09"),"ano" ] = 2021
df.rename(columns={'ano':'temporada'},inplace=True)
```

```python
df = df[df['temporada']>=2018]

# Verificando que não há dados NAN no dataset
df.isna().sum()
```

```python
#Definindo colunas de vitorias e derrotas e empates mandante/visitante
df['win_mandante'] = df.apply(lambda x: 1 if x["vencedor"] == x["mandante"] else 0,axis=1)
df['loss_mandante'] = df.apply(lambda x: 1 if x["vencedor"] == x["visitante"] else 0,axis=1)
df['win_visitante'] = df.apply(lambda x: 1 if x["vencedor"] == x["visitante"] else 0,axis=1)
df['loss_visitante'] = df.apply(lambda x: 1 if x["vencedor"] == x["mandante"] else 0,axis=1)
df['empate'] = df.apply(lambda x: 1 if x["vencedor"] == '-' else 0,axis=1)
```

```python
#df2018_2022['chutes'] = df2018_2022['chutes'].astype('Int64')

df['posse_de_bola_mandante'] = df['posse_de_bola_mandante'].replace({'%': ''}, regex=True)
df['precisao_passes_mandante'] = df['precisao_passes_mandante'].replace({'%': ''}, regex=True)
df['posse_de_bola_visitante'] = df['posse_de_bola_visitante'].replace({'%': ''}, regex=True)
df['precisao_passes_visitante'] = df['precisao_passes_visitante'].replace({'%': ''}, regex=True)
```

```python
# Substitui valores ausentes ou 'None' por NaN
df['posse_de_bola_mandante'] = df['posse_de_bola_mandante'].replace({'None': np.nan})
df['precisao_passes_mandante']= df['precisao_passes_mandante'].replace({'None': np.nan})
df['posse_de_bola_visitante'] = df['posse_de_bola_visitante'].replace({'None': np.nan})
df['precisao_passes_visitante'] = df['precisao_passes_visitante'].replace({'None': np.nan})

# convertendo o percentual para numero
df['posse_de_bola_mandante'] = df['posse_de_bola_mandante'].astype('Int64') /100
df['precisao_passes_mandante'] = df['precisao_passes_mandante'].astype('Int64') /100
df['posse_de_bola_visitante'] = df['posse_de_bola_visitante'].astype('Int64') /100
df['precisao_passes_visitante'] = df['precisao_passes_visitante'].astype('Int64') /100
```
